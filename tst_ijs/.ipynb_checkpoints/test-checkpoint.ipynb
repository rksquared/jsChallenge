# Let's play with functions


```javascript
function strMult(strInp, intOperator) {
    var result = Number(strInp)*intOperator;
    return result;
}
```

```javascript
var five = strMult('5', 5)
five
```

If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of these multiples is 23.

Find the sum of all the multiples of 3 or 5 below 1000.

```javascript
function euler1() {
    var sumList = []
    for (i=0; i <= 1000; i++){
        if (i % 3 === 0 | i % 5 === 0) {
            sumList.push(i);
        };
    };
    return sumList.reduce(function(x,y){return x+y;});
}
```

```javascript
euler1()
```
