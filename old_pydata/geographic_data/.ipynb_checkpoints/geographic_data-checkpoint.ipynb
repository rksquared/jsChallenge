```python
import pandas as pd
import matplotlib.pyplot as plt
import mpl_toolkits.basemap as Basemap
```

```python
airlines = pd.read_csv("airlines.csv")
airports = pd.read_csv("airports.csv")
routes = pd.read_csv("routes.csv")
```

### Convert spherical longitude and latitude coordinate values into cartesian coordinate values

Take the values from the the 'longitude' and 'latitude' columns in the airports dataframe, convert them to lists (the constructor only accepts lists, not pd.series) and convert to cartesian. Print out and confirm

```python
m = Basemap(projection='merc', llcrnrlat=-80, urcrnrlat=80, llcrnrlon=-180, urcrnrlon=180)

longitude = airports['longitude'].tolist()
latitude = airports['latitude'].tolist()

x,y=m(longitude, latitude)

print(x)

print(y)
```

### Display a basemap.scatter() of the newly converted cartesian long/lat lists 

```python
m.scatter(x,y,s=1)

plt.show()
```

Let's display coastlines

```python
m.scatter(x, y, s=1)
m.drawcoastlines()
plt.show()
```

### Tweaking the basemap visualization with matplotlib 

Recall that basemap used matplotlib under the hood. We can modify/tweak the basemap visualization the same way we modified the matplot and seaborn plots.

Let's scale/size the figure and add a title to our geo-scatter plot

```python
fig, ax = plt.subplots(figsize=(15,20))
plt.title("Scaled Up Earth With Coastlines")

m = Basemap(projection='merc', llcrnrlat=-80, urcrnrlat=80, llcrnrlon=-180, urcrnrlon=180)
longitudes = airports["longitude"].tolist()
latitudes = airports["latitude"].tolist()
x, y = m(longitudes, latitudes)

m.scatter(x, y, s=1)
m.drawcoastlines()
plt.show()
```
